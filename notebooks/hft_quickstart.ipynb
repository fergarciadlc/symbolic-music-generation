````python
from transformers import pipeline
````

````python
sentiment_analysis_pipeline = pipeline("sentiment-analysis")
data = [
    "I love you",
    "I hate you"
]
sentiment_analysis_pipeline(data)
````

Note that when we don't specify the model, hft will use a default one if available.

In this case for `sentiment-analysis`, the library is using `distilbert/distilbert-base-uncased-finetuned-sst-2-english` (from the log above)

--- 
Model works! But does it perform well in other languages?

````python
data = ["I love you", "I hate you", "Mi estimado", "Je t'aime", "æ„›ã—ã¦ã¾ã™"]
results = sentiment_analysis_pipeline(data)
for sentence, prediction in zip(data, results):
    print(f"{sentence} --> {prediction}")
````

A bit odd innit?

Let's try this model: https://huggingface.co/tabularisai/multilingual-sentiment-analysis

````python
multilingual_pipe = pipeline("text-classification", model="tabularisai/multilingual-sentiment-analysis")
````

````python
results = multilingual_pipe(data)
for sentence, prediction in zip(data, results):
    print(f"{sentence} --> {prediction}")
````
